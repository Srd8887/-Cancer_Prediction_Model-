```python
df = df.drop("Unnamed: 32", axis=1 )
```

```python
df.head()
```

```python
sns.heatmap(df.isnull())
```

```python
df.drop('id',axis=1,inplace=True)
```

```python
sns.heatmap(df.isnull())
```

```python
df["diagnosis"] = [1 if value == "M" else 0 for value in df["diagnosis"]]
```

```python
df.head()
```

```python
df['diagnosis'] = df['diagnosis'].astype("category",copy=False)
```

```python
df.info()
```

```python
df['diagnosis'].value_counts().plot(kind='bar')
```

```python
# divide into target variable and predictors
x = df.drop(['diagnosis'],axis=1)
y = df['diagnosis']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()# create object
```

```python
X_scaled = scaler.fit_transform(x)
```

```python
print(X_scaled)
```

```python
# Spilit the data
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.3, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
# Train the model

from sklearn.linear_model import LogisticRegression

Model = LogisticRegression()
```

```python
Model.fit(x_train,y_train)
```

```python
y_pred = Model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
# Evaluation of the Model

from sklearn.metrics import accuracy_score,precision_score,recall_score,r2_score,roc_auc_score,confusion_matrix,classification_report
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
precision = precision_score(y_test,y_pred)
precision
```

```python
recall =recall_score(y_test,y_pred)
recall
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
roc_auc = roc_auc_score(y_test,y_pred)
roc_auc
```

```python
Con_Metrix = confusion_matrix(y_test,y_pred)
Con_Metrix
```

```python
Report = classification_report(y_test,y_pred)
print(Report)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Calculate FPR, TPR
fpr, tpr, thresholds = roc_curve(y_test, y_pred)

# Calculate AUC
roc_auc = roc_auc_score(y_test, y_pred)

# Plot ROC Curve
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0,1], [0,1], color='red', linestyle='--', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```
